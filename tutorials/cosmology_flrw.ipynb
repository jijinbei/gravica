# FLRW Cosmology — Stress-Energy Tensor and Cosmological Constant

The Friedmann-Lemaitre-Robertson-Walker (FLRW) metric models a homogeneous, isotropic universe.
This tutorial covers:

1. Computing the Ricci tensor and Ricci scalar for FLRW
2. Deriving the Einstein tensor
3. Constructing the stress-energy tensor with a cosmological constant
4. Computing the Schouten tensor

$$
ds^2 = dt^2 - a(t)^2 \left[\frac{dr^2}{1 - kr^2} + r^2\,d\Omega^2\right]
$$

```python
import os
from symbolica import set_license_key, S

_key = os.environ.get("SYMBOLICA_LICENSE", "")
if _key:
    set_license_key(_key)

from gravica import (
    ChristoffelSymbols,
    RiemannTensor,
    RicciTensor,
    EinsteinTensor,
    SchoutenTensor,
    StressEnergyTensor,
    ricci_scalar,
    display,
)
from gravica.simplify import simplify, str_is_zero
from gravica.metrics import flrw
```

## 1. FLRW Metric

Parameters:
- $k$: spatial curvature ($k = +1, 0, -1$ for closed, flat, or open)
- $a(t)$: scale factor describing the expansion of the universe

```python
g = flrw()
coords = ["t", "r", "\\theta", "\\varphi"]

print("FLRW metric:")
display.components_table(display.nonzero_components(g, coords), g)
```

```python
christoffel = ChristoffelSymbols(g)
riemann = RiemannTensor(christoffel)
ricci = RicciTensor(riemann)
einstein = EinsteinTensor(ricci)
```

## 2. Christoffel Symbols

```python
display.components_table(display.nonzero_components(christoffel, coords), christoffel)
```

## 3. Ricci Tensor and Ricci Scalar

FLRW spacetime is **not** a vacuum solution, so $R_{ab} \neq 0$.
The results involve the scale factor $a(t)$ and its time derivatives.

```python
print("Ricci tensor R_{ab}:")
display.components_table(display.nonzero_components(ricci, coords), ricci)

R = simplify(ricci_scalar(ricci))
print(f"\nRicci scalar R = {R}")
```

## 4. Einstein Tensor

$$G_{ab} = R_{ab} - \frac{1}{2}g_{ab}R$$

This is the geometric side of the Friedmann equations.

```python
print("Einstein tensor G_{ab}:")
display.components_table(display.nonzero_components(einstein, coords), einstein)
```

## 5. Stress-Energy Tensor with Cosmological Constant

The Einstein field equations with cosmological constant:

$$G_{ab} + \Lambda g_{ab} = 8\pi G\,T_{ab}$$

`StressEnergyTensor` computes $8\pi G\,T_{ab} = G_{ab} + \Lambda g_{ab}$.

```python
from symbolica import PrintMode


def _greek(name):
    tex = f"\\{name}"

    def _print(expr, mode, **kw):
        if mode == PrintMode.Latex:
            return tex

    return S(name, custom_print=_print)


Lambda = _greek("Lambda")

stress_energy = StressEnergyTensor(einstein, cosmological_constant=Lambda)

print("8*pi*G * T_{ab} (with cosmological constant):")
display.components_table(display.nonzero_components(stress_energy, coords), stress_energy)
```

## 6. Schouten Tensor

The Schouten tensor plays a key role in conformal geometry:

$$S_{ab} = \frac{1}{n-2}\left(R_{ab} - \frac{R}{2(n-1)}g_{ab}\right)$$

```python
schouten = SchoutenTensor(ricci)

print("Schouten tensor S_{ab}:")
display.components_table(display.nonzero_components(schouten, coords), schouten)
```
