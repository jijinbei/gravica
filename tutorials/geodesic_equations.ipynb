# Geodesic Equations — Equations of Motion in Curved Spacetime

A free particle (subject only to gravity) follows a **geodesic** in curved spacetime.
The geodesic equation is:

$$
\frac{d^2 x^a}{d\tau^2} + \Gamma^a_{bc}\,\frac{dx^b}{d\tau}\,\frac{dx^c}{d\tau} = 0
$$

Gravica's `geodesic_equations()` generates the left-hand side of this equation symbolically for each coordinate.

```python
import os
from symbolica import set_license_key

_key = os.environ.get("SYMBOLICA_LICENSE", "")
if _key:
    set_license_key(_key)

from gravica import ChristoffelSymbols, geodesic_equations
from gravica.simplify import simplify
from gravica.metrics import schwarzschild
```

## 1. Schwarzschild Geodesics

Deriving the equations of motion for a particle around a Schwarzschild black hole.

```python
g = schwarzschild()
christoffel = ChristoffelSymbols(g)
geo = geodesic_equations(christoffel)

coord_names = ["t", "r", "theta", "phi"]

print("Velocity symbols (dx^a/dtau):")
for name, v in zip(coord_names, geo.velocities):
    print(f"  d{name}/dtau = {v}")

print()
print("Acceleration symbols (d^2 x^a/dtau^2):")
for name, a in zip(coord_names, geo.accelerations):
    print(f"  d^2 {name}/dtau^2 = {a}")
```

## 2. Displaying the Geodesic Equations

Each equation is the left-hand side set equal to zero.

```python
for name, eq in zip(coord_names, geo.equations):
    print(f"--- {name} component ---")
    print(f"{simplify(eq)} = 0")
    print()
```

## 3. Physical Interpretation

- **$t$ component**: time-translation symmetry implies a conserved energy
- **$r$ component**: radial motion — strong gravitational effects near $r_s$
- **$\theta$ component**: motion in the equatorial plane ($\theta = \pi/2$) is stable
- **$\varphi$ component**: axial symmetry implies a conserved angular momentum

## 4. Kerr Geodesics

For a rotating black hole, frame dragging introduces $t$–$\varphi$ coupling terms.

```python
from gravica.metrics import kerr

g_kerr = kerr()
christoffel_kerr = ChristoffelSymbols(g_kerr)
geo_kerr = geodesic_equations(christoffel_kerr)

for name, eq in zip(coord_names, geo_kerr.equations):
    print(f"--- {name} component ---")
    print(f"{simplify(eq)} = 0")
    print()
```
