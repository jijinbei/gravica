# Predefined Metrics

Gravica ships with 8 built-in spacetime metrics commonly used in general relativity.
This tutorial showcases each one.

| Metric | Function | Physics |
|--------|----------|---------|
| Minkowski | `minkowski()` | Flat spacetime |
| Schwarzschild | `schwarzschild()` | Static black hole |
| Kerr | `kerr()` | Rotating black hole |
| FLRW | `flrw()` | Cosmology |
| Gödel | `godel()` | Rotating universe |
| Reissner–Nordström | `reissner_nordstrom()` | Charged black hole |
| de Sitter | `de_sitter()` | Positive cosmological constant |
| Anti-de Sitter | `anti_de_sitter()` | Negative cosmological constant |

````python
import os
from symbolica import set_license_key

_key = os.environ.get("SYMBOLICA_LICENSE", "")
if _key:
    set_license_key(_key)

from gravica import display
from gravica.metrics import (
    minkowski,
    schwarzschild,
    kerr,
    flrw,
    godel,
    reissner_nordstrom,
    de_sitter,
    anti_de_sitter,
)
````

## 1. Minkowski Metric — Flat Spacetime

The arena of special relativity. Zero curvature everywhere.

$$
ds^2 = dt^2 - dx^2 - dy^2 - dz^2
$$

````python
g = minkowski()
coords = ["t", "x", "y", "z"]

print(f"Coordinates: {g.coords}")
print(f"Dimension:   {g.dim}")
print(f"Diagonal:    {g.is_diagonal()}")
print()
display.components_table(display.nonzero_components(g, coords), g)
````

## 2. Schwarzschild Metric — Static Black Hole

Exterior spacetime of a spherically symmetric, static mass $M$.
$r_s = 2GM/c^2$ is the Schwarzschild radius.

$$
ds^2 = \frac{r - r_s}{r}\,dt^2 - \frac{r}{r - r_s}\,dr^2 - r^2\,d\Omega^2
$$

````python
g = schwarzschild()
coords = ["t", "r", "\\theta", "\\varphi"]

print(f"Diagonal: {g.is_diagonal()}")
display.components_table(display.nonzero_components(g, coords), g)
````

## 3. Kerr Metric — Rotating Black Hole

A black hole with mass $M$ and angular momentum $J = Ma$ in Boyer–Lindquist coordinates.

$$
\Sigma = r^2 + a^2 \cos^2\theta, \quad \Delta = r^2 - 2Mr + a^2
$$

The off-diagonal component $g_{t\varphi}$ encodes **frame dragging**.

````python
g = kerr()
coords = ["t", "r", "\\theta", "\\varphi"]

print(f"Diagonal: {g.is_diagonal()}")  # False — off-diagonal t-phi component
display.components_table(display.nonzero_components(g, coords), g)
````

## 4. FLRW Metric — Cosmology

The homogeneous, isotropic universe. Scale factor $a(t)$ and spatial curvature $k$.

$$
ds^2 = dt^2 - a(t)^2 \left[\frac{dr^2}{1 - kr^2} + r^2\,d\Omega^2\right]
$$

````python
g = flrw()
coords = ["t", "r", "\\theta", "\\varphi"]

display.components_table(display.nonzero_components(g, coords), g)
````

## 5. Gödel Metric — Rotating Universe

Kurt Gödel's exact solution containing closed timelike curves.

````python
g = godel()
coords = ["t", "x", "y", "z"]

print(f"Diagonal: {g.is_diagonal()}")
display.components_table(display.nonzero_components(g, coords), g)
````

## 6. Reissner–Nordström Metric — Charged Black Hole

A non-rotating black hole with electric charge $Q$.

$$
f(r) = 1 - \frac{r_s}{r} + \frac{r_Q^2}{r^2}
$$

````python
g = reissner_nordstrom()
coords = ["t", "r", "\\theta", "\\varphi"]

display.components_table(display.nonzero_components(g, coords), g)
````

## 7. de Sitter Metric — Positive Cosmological Constant

Describes an accelerating, expanding universe with $\Lambda > 0$.

$$
ds^2 = \left(1 - \frac{\Lambda r^2}{3}\right)dt^2 - \left(1 - \frac{\Lambda r^2}{3}\right)^{-1}dr^2 - r^2\,d\Omega^2
$$

````python
g = de_sitter()
coords = ["t", "r", "\\theta", "\\varphi"]

display.components_table(display.nonzero_components(g, coords), g)
````

## 8. Anti-de Sitter Metric — Negative Cosmological Constant

Central to the AdS/CFT correspondence in theoretical physics.

````python
g = anti_de_sitter()
coords = ["t", "r", "\\theta", "\\varphi"]

display.components_table(display.nonzero_components(g, coords), g)
````

## Summary

All predefined metrics can be fed directly into the tensor pipeline via `from_metric`:

```python
from gravica import RicciTensor
from gravica.metrics import schwarzschild

ricci = RicciTensor.from_metric(schwarzschild())
```
